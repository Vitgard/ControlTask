```python
penguins = sns.load_dataset("penguins")
penguins.dropna(inplace=True)
penguins.head()
```

```python
penguins.isnull().sum()
```

```python
penguins.shape
```

```python
sns.scatterplot(data=penguins.dropna(), x='body_mass_g', y='flipper_length_mm', hue='sex', style='island', size='bill_depth_mm')
```

```python
first=['body_mass_g', 'flipper_length_mm', 'sex', 'bill_depth_mm']
g=sns.PairGrid(penguins[first])
g.map(sns.scatterplot)
```

```python
penguins.corr()
```

```python
sns.heatmap(penguins.corr(), vmin=-1, vmax=1)
```
